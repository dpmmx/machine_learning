### Example 1 : Plotting Sign Curve using Matplot Library

#### 1. Basic example of Matplot Library ploting Sign Curve.
#### 2. Also used Labeling on X-Axis, Y-Axis and Title 

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
import math
```

```python
x = np.arange(0, math.pi*2, 0.05)
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.xlabel("Angle")
plt.ylabel("Sine")
plt.title("Sine Wave")
plt.show()
```

### Example 2 - PyLab plotting
PyLab is a procedural interface to the Matplotlib object-oriented plotting library. Matplotlib is the whole package; matplotlib.pyplot is a module in Matplotlib; and PyLab is a module that gets installed alongside Matplotlib.

PyLab is a convenience module that bulk imports matplotlib.pyplot (for plotting) and NumPy (for Mathematics and working with arrays) in a single name space. Although many examples use PyLab, it is no longer recommended.

```python
from pylab import *
```

```python
x = linspace(-3, 3, 20)
```

```python
y = x**2
```

```python
plot(x, y)
show()
```

Plots can be overlaid. Just use the multiple plot commands. Use <b>clf()</b> to clear the plot.

```python
plot(x, sin(x))
plot(x, cos(x), 'r-')
plot(x, -sin(x), 'g--')
show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
import math
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)
fig = plt.figure(facecolor='c',edgecolor='w',figsize=(2,2),dpi=500)
ax = fig.add_axes([0,0,1,1]) # left, bottom, width, height
ax.plot(x,y)
ax.set_title("Sine Wave")
ax.set_xlabel("Angle in Radian")
ax.set_ylabel("Value of Sine")
plt.show()
```

```python
import matplotlib.pyplot as plt

y = [0, 10, 20, 30, 40, 50, 60, 70]
x1 = [2, 16, 30, 42, 55, 68, 77, 88]
x2 = [5, 6, 12, 18, 28, 40, 52, 65]

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

l1 = ax.plot(x1,y,'ys-') # solid line with yellow colour and square marker
l2 = ax.plot(x2,y,'go--') # dash line with green colour and circle marker

ax.legend(labels = ('TV', 'Smartphone'), loc = 'lower right') # legend placed at lower right
ax.set_title("Advertisement Effect on Sales", loc="center")
ax.set_xlabel('Medium')
ax.set_ylabel('Sales')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math
x = np.arange(0, math.pi*2, 0.05)
fig=plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.55, 0.55, 0.3, 0.3]) # inset axes
y = np.sin(x)
axes1.plot(x, y, 'b')
axes2.plot(x,np.cos(x),'r')
axes1.set_title('sine')
axes2.set_title("cosine")
plt.show()
```
